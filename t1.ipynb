```python
import pandas as p
import matplotlib.pyplot as m
import seaborn as s
data=p.read_csv('C:\\Users\\ARJUN\\OneDrive\\Desktop\\mtcars.csv')

mpg=data['mpg']
m.hist(mpg,bins='auto',color='k',edgecolor='g')
m.xlabel('miles per gallon(mpg)')
m.ylabel('frequency')
m.title('frequency distribution of variables')
m.show()

wt=data['wt']
iv=range(len(data))
m.scatter(iv,mpg,color='k',label='mpg')
m.scatter (iv,wt,color='g',label='wt')
m.title('relationship b/w weight & mpg')
m.legend()
m.show()

c=data['am'].value_counts()
color=['k','g']
m.bar(c.index,c.values,color='k',width=0.3)
m.xticks([0,1],['0-automatic','1-manual'])
m.xlabel('transmission type')
m.ylabel('no of cars')
m.title('frequency distributin of transmission type of car')
m.show()

s.boxplot(mpg,color='blue')
m.xlabel('mpg')
m.ylabel('values')
m.title('box plot for mpg &values')
```
